```python
for i in range(1,512):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=6ef6a75eb8d77bff00563bc0ce73c51d&language=en-US&page={}'.format(i))
    if response.status_code == 200 and 'results' in response.json():
        temp_df = pd.DataFrame(response.json()['results'])[['title', 'id', 'genre_ids', 'overview']]
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Skipping page {i}: {response.status_code}")
```

```python
response = requests.get('https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US')
genres = (response.json()['genres'])
```

```python
genres
```

```python
def get_genre_string(genre_ids_list, genres):
    if not genre_ids_list:
        return "No genres"

    # Create a mapping from id to name
    genre_dict = {g['id']: g['name'] for g in genres}

    # Collect names for the given genre_ids
    genre_names = [genre_dict.get(gid) for gid in genre_ids_list if gid in genre_dict]

    # Return comma-separated string
    return ', '.join(genre_names) if genre_names else "No genres"
```

```python
df['genre'] = df['genre_ids'].apply(lambda x: get_genre_string(x, genres))
df.head()
```

```python
df.drop('genre_ids', axis=1, inplace=True)
df.head()
```

```python
df.drop('id', axis=1, inplace=True)
df.head()
```
